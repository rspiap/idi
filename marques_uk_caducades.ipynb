```python
from google.colab import files
import pandas as pd

# Demana a l'usuari que pugi un arxiu Excel
uploaded = files.upload()

# Obtenir el nom de l'arxiu pujat
nom_arxiu = list(uploaded.keys())[0]

# Carregar l'arxiu Excel a partir de la fila 14
df = pd.read_excel(nom_arxiu, skiprows=13)
```

```python
import datetime

avui = pd.to_datetime(datetime.date.today())

df['data'] = pd.to_datetime(df['Data fi vigència'], errors='coerce')
```

```python
# Filtre 1: Estat = "Alta"
filtre_estat = df['Estat'] == 'Alta'

# Filtre 2: Catàleg = "Marca"
filtre_catalog = df['Catàleg'] == 'Marca'

# Filtre 3: Registre d'inscripció = "Oficina de Propietat Intel·lectual de la Unió Europea (EUIPO)"
filtre_registre = df['Registre d\'inscripció'] == 'Altres registres'
filtre_registre = df['Codi registre'].str.startswith("UK")

# Filtre 4: Data Fi vigència anterior a la data d'avui
filtre_vigencia = df['Data fi vigència'] < avui

# Aplica els filtres
df_filtrat = df[filtre_estat & filtre_catalog & filtre_registre & filtre_vigencia]

# Mostra el DataFrame filtrat
print(len(df_filtrat))
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
from bs4 import BeautifulSoup
import json

def euipo(codi):

    url = "https://branddb.wipo.int/en/quicksearch/brand/GB50" + codi

    driver.get(url)

    try:
        # Espera fins que almenys un element amb la classe que contingui "col-33" estigui present
        elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li[class*='col-33']"))
        )

        # Crea un diccionari per emmagatzemar els resultats
        result_dict = {}

        # Imprimeix el text de tots els elements trobats
        for element in elements:
          soup = BeautifulSoup(element.get_attribute("outerHTML"), 'html.parser')
          # Extreu tots els elements amb la classe "label"
          label_elements = soup.select('.label')
          content_elements = soup.select("[class*=content]")

          # Itera sobre els elements amb la classe "label" i "content" simultàniament
          for label_element, content_element in zip(label_elements, content_elements):
              # Extreu el text de cada element i afegeix-lo al diccionari
              label_text = label_element.text.strip()
              content_text = content_element.text.strip()
              result_dict[label_text] = content_text

        result_dict["Url"] = url

        # Imprimeix el diccionari
        json_result = json.dumps(result_dict, indent=2)
        return json_result

    except:
        print("ERROR")
```

```python
json_final = {"euipo": []}

l = len(df_filtrat)
i = 1

# Itera a través de la columna "Codi registre" i obté la primera fila
for x in df_filtrat["Codi registre"]:
  print(i, "de", l)
  resultat_iteracio = euipo(x)
  json_final["euipo"].append(resultat_iteracio)
  i += 1

# Mostra el JSON final
print(json_final)
```

```python
# Filtra els elements nuls de la llista
filtered_data = [item for item in json_final["euipo"] if item is not None]

# Converteix la llista de cadenes JSON a una llista d'objectes Python
json_objects = [json.loads(item) for item in filtered_data]

# Passar aquesta llista d'objectes Python directament a Pandas
import pandas as pd

euipo_final = pd.json_normalize(json_objects)
euipo_final
```
